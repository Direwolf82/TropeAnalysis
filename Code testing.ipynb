```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as BS

def tropes_on_page(page_address):
    page = requests.get(page_address)
    soup = BS(page.content, 'html.parser')
    rows = list(soup.findAll('tr'))
    tropes = list()
    paired_elements = list()
    types = list()
    comment = list()
    for element in rows:
        first = list(element.find('div').contents)[0]
        second = ''
        if first == 'type':
            second = list(element.findAll('a'))
            second = list(second[len(second)-1].children)[0]
            types.append(second)
            paired_elements.append([first,second])
        elif first == 'comment':
            second = element.findAll('i')[0]
            second = second.text
            comment.append(second)
            paired_elements.append([first,second])
    df = pd.DataFrame({'type':types, 'comment':comment})
    return df
```
